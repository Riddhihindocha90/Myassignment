**2. Suitable graph of the mean BMI (BMI not body_mass weight_in_kg/height(m2)) of female and male penguins. (Assume the height is the flipper length)**  (5 marks)


```python
#Your answer here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'penguins_df'
# Replace 'penguins.csv' with your actual dataset
penguins_df = pd.read_csv('C:/Users/Riddhi/Downloads/Myassignment/data/penguins.csv')

# Calculate BMI for each penguin
penguins_df['BMI'] = penguins_df['body_mass_g'] / ((penguins_df['flipper_length_mm'] / 1000) ** 2)

# Group by gender and calculate mean BMI
mean_bmi_by_gender = penguins_df.groupby('sex')['BMI'].mean().reset_index()

# Plotting the mean BMI for female and male penguins
sns.barplot(x='sex', y='BMI', data=mean_bmi_by_gender)
plt.title("Mean BMI of Female and Male Penguins")
plt.xlabel("Gender")
plt.ylabel("Mean BMI")
plt.show()
```

Convince the museum board 

The Firehouse Museum claims they ran an event in 2014 that brought an incredible number of visitors, and that they should get extra budget to run a similar event again.  The other museums think these types of events aren't that important, and budgets should be split purely based on recent visitors on an average day.  

To show the museum board how the event compared to regular traffic at each museum, create a suitable chart that shows how the number of visitors to each museum evolved over time.
(4 marks)

```python
#Your solution here
#Your solution here

import pandas as pd
import matplotlib.pyplot as plt

# Load the museum visitors data 
file_path = "C:/Users/Riddhi/Downloads/Myassignment/data/museum_visitors.csv"
museum_df = pd.read_csv(file_path)

# Check the structure of the dataset
print(museum_df.head())

# Convert the 'Date' column to datetime format
museum_df['Date'] = pd.to_datetime(museum_df['Date'])

# Create a line chart to show the evolution of visitors over time for each museum
plt.figure(figsize=(12, 6))

# Iterate through each museum and plot the number of visitors over time
for museum in museum_df.columns[1:]:
    plt.plot(museum_df['Date'], museum_df[museum], label=museum)

# Highlight the year 2014 for emphasis on the event
plt.axvline(pd.to_datetime('2014-01-01'), color='red', linestyle='--', label='Event Year (2014)')

# Set plot title and labels
plt.title('Evolution of Museum Visitors Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Visitors')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

```
